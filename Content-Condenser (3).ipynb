```python
import tkinter as tk
from tkinter import scrolledtext, messagebox, END, Menu, filedialog, colorchooser
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from collections import Counter
from heapq import nlargest
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load spaCy model globally to avoid reloading it every time the button is pressed
nlp = spacy.load('en_core_web_sm')

def summarize_text():
    # Get text from the text_box
    text = text_box.get("1.0", "end-1c")
    
    if not text.strip():
        messagebox.showerror("Error", "Please enter some text to condense.")
        return

    # Tokenization and removing stopwords
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc 
              if not token.is_stop and not token.is_punct and token.text != '\n']

    # Calculating word frequency
    word_freq = Counter(tokens)
    if not word_freq:
        messagebox.showerror("Error", "No meaningful words found in the text.")
        return
    
    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word] / max_freq
    
    # Sentence tokenization
    sent_token = [sent.text for sent in doc.sents]

    sent_score = {}
    for sent in sent_token:
        for word in sent.split():
            if word.lower() in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word]
                else:
                    sent_score[sent] += word_freq[word]

    # Select top-scoring sentences based on user input
    try:
        num_sentences = int(num_sentences_entry.get())
    except ValueError:
        messagebox.showerror("Error", "Number of sentences must be an integer.")
        return

    if num_sentences <= 0:
        messagebox.showerror("Error", "Number of sentences must be greater than zero.")
        return

    summarized_sentences = nlargest(num_sentences, sent_score, key=sent_score.get)
    if not summarized_sentences:
        messagebox.showerror("Error", "Could not generate a summary with the given number of sentences.")
        return

    # Display summarized text in the result_box
    result_box.delete(1.0, END)
    result_box.insert(END, " ".join(summarized_sentences))
    
    # Update counts
    update_counts()

def clear_text():
    text_box.delete(1.0, END)
    result_box.delete(1.0, END)
    num_sentences_entry.delete(0, END)
    num_sentences_entry.insert(END, "3")  # Reset to default value
    update_counts()

def show_about():
    messagebox.showinfo("About", "Content Condenser\nVersion 1.0\nPowered by spaCy and Tkinter")

def save_summary():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt",
                                             filetypes=[("Text files", "*.txt"),
                                                        ("All files", "*.*")])
    if file_path:
        with open(file_path, 'w') as file:
            file.write(result_box.get(1.0, END))
        messagebox.showinfo("Saved", f"Summary saved to {file_path}")

def load_text():
    file_path = filedialog.askopenfilename(filetypes=[("Text files", "*.txt"),
                                                      ("All files", "*.*")])
    if file_path:
        with open(file_path, 'r') as file:
            text = file.read()
            text_box.delete(1.0, END)
            text_box.insert(END, text)
        update_counts()

def copy_to_clipboard():
    root.clipboard_clear()
    root.clipboard_append(result_box.get(1.0, END))
    messagebox.showinfo("Copied", "Summary copied to clipboard")

def update_counts():
    input_text = text_box.get("1.0", "end-1c")
    output_text = result_box.get("1.0", "end-1c")

    input_char_count = len(input_text)
    input_word_count = len(input_text.split())

    output_char_count = len(output_text)
    output_word_count = len(output_text.split())

    input_count_label.config(text=f"Input - Characters: {input_char_count}, Words: {input_word_count}")
    output_count_label.config(text=f"Output - Characters: {output_char_count}, Words: {output_word_count}")

def generate_word_cloud():
    text = text_box.get("1.0", "end-1c")
    if not text.strip():
        messagebox.showerror("Error", "Please enter some text to generate a word cloud.")
        return
    
    # Customizing the appearance of the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis',
                          max_words=100, contour_color='black', contour_width=3,
                          random_state=42).generate(text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

def customize_color():
    color = colorchooser.askcolor()[1]
    if color:
        text_box.config(bg=color)
        result_box.config(bg=color)

# GUI setup
root = tk.Tk()
root.title("Content Condenser")
root.geometry("800x600")
root.configure(bg="#f0f0f0")

# Menu bar setup
menu_bar = Menu(root)
root.config(menu=menu_bar)

# Add File menu
file_menu = Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="Load Text", command=load_text)
file_menu.add_command(label="Save Summary", command=save_summary)
file_menu.add_separator()
file_menu.add_command(label="Clear", command=clear_text)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)

# Add Edit menu
edit_menu = Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Edit", menu=edit_menu)
edit_menu.add_command(label="Copy Summary to Clipboard", command=copy_to_clipboard)
# edit_menu.add_command(label="Select Font", command=select_font)  # Removed font selection option
edit_menu.add_command(label="Customize Background Color", command=customize_color)

# Add Tools menu
tools_menu = Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Tools", menu=tools_menu)
tools_menu.add_command(label="Generate Word Cloud", command=generate_word_cloud)

# Add Help menu
help_menu = Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="Help", menu=help_menu)
help_menu.add_command(label="About", command=show_about)

# Layout frames
input_frame = tk.Frame(root, bg="#f0f0f0")
input_frame.pack(pady=10, padx=10, fill=tk.BOTH, expand=True)

controls_frame = tk.Frame(root, bg="#f0f0f0")
controls_frame.pack(pady=5, padx=10, fill=tk.X)

output_frame = tk.Frame(root, bg="#f0f0f0")
output_frame.pack(pady=10, padx=10, fill=tk.BOTH, expand=True)

counts_frame = tk.Frame(root, bg="#f0f0f0")
counts_frame.pack(pady=5, padx=10, fill=tk.X)

# Text box for input
text_box_label = tk.Label(input_frame, text="Enter Text:", bg="#f0f0f0", font=("Helvetica", 12))
text_box_label.pack(anchor='w', padx=10, pady=(10, 0))
text_box = scrolledtext.ScrolledText(input_frame, width=80, height=10, wrap=tk.WORD, font=("Helvetica", 11))
text_box.pack(padx=10, pady=5, fill=tk.BOTH, expand=True)
text_box.bind("<<Modified>>", lambda e: update_counts())

# Entry field for the number of sentences
num_sentences_label = tk.Label(controls_frame, text="Number of Sentences:", bg="#f0f0f0", font=("Helvetica", 12))
num_sentences_label.grid(row=0, column=0, padx=5, pady=5)
num_sentences_entry = tk.Entry(controls_frame, width=10, font=("Helvetica", 11))
num_sentences_entry.insert(END, "3")  # Default value
num_sentences_entry.grid(row=0, column=1, padx=5, pady=5)

# Buttons for summarizing and clearing
summarize_button = tk.Button(controls_frame, text="Condense", command=summarize_text, bg="#4caf50", fg="white", font=("Helvetica", 11))
summarize_button.grid(row=0, column=2, padx=5, pady=5)
clear_button = tk.Button(controls_frame, text="Clear", command=clear_text, bg="#f44336", fg="white", font=("Helvetica", 11))
clear_button.grid(row=0, column=3, padx=5, pady=5)

# Result box for output
result_box_label = tk.Label(output_frame, text="Condensed Text:", bg="#f0f0f0", font=("Helvetica", 12))
result_box_label.pack(anchor='w', padx=10, pady=(10, 0))
result_box = scrolledtext.ScrolledText(output_frame, width=80, height=10, wrap=tk.WORD, font=("Helvetica", 11))
result_box.pack(padx=10, pady=5, fill=tk.BOTH, expand=True)
result_box.bind("<<Modified>>", lambda e: update_counts())

# Labels for character and word count
input_count_label = tk.Label(counts_frame, text="Input - Characters: 0, Words: 0", bg="#f0f0f0", font=("Helvetica", 10))
input_count_label.pack(anchor='w', padx=10, pady=(5, 0))
output_count_label = tk.Label(counts_frame, text="Output - Characters: 0, Words: 0", bg="#f0f0f0", font=("Helvetica", 10))
output_count_label.pack(anchor='w', padx=10, pady=(5, 0))

root.mainloop()
```
